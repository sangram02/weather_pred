```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import GaussianNB
data = pd.read_csv("./weather.csv")
data
```

```python
# data.loc[data['Play'] == 'no','Play'] = 0
# data.loc[data['Play'] == 'yes','Play'] = 1
```

```python
input = data.drop(['Play'],axis=1)
output = data['Play']
output
```

```python
numerics = LabelEncoder()
input['Outlook_n'] = numerics.fit_transform(input['Outlook'])
input['Temp_n'] = numerics.fit_transform(input['Temp'])
input['Humidity_n'] = numerics.fit_transform(input['Humidity'])
input['Windy_n'] = numerics.fit_transform(input['Windy'])
input
```

```python
input = input.drop(['Outlook','Temp','Humidity','Windy'],axis=1)
input
```

```python
output
```

```python
classifier = GaussianNB()
classifier.fit(input,output)
classifier.score(input,output)
```
